```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits = pd.read_csv('data/tmdb_5000_credits.csv')
movies_df = pd.read_csv('data/tmdb_5000_movies.csv')
```

```python
credits.head(5)
```

```python
movies_df.head()
```

```python
print('Credits:',credits.shape)
print('Movies_DF:',movies_df.shape)
```

```python
credits_col_rename = credits.rename(index = str, columns={'movie_id':'id'})
movies_df_merge = movies_df.merge(credits_col_rename, on = 'id')
```

```python
movies_df_merge.head()
```

```python
movie_clean_df = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
```

```python
movie_clean_df.head()
```

```python
movie_clean_df.info()
```

```python
# Calculating all the components based on the above formula
v = movie_clean_df['vote_count']
R = movie_clean_df['vote_average']
C = movie_clean_df['vote_average'].mean()
m = movie_clean_df['vote_count'].quantile(0.7)
```

```python
movie_clean_df['weighted_average'] = ((R*v) + (C*m)) / (v+m)
```

```python
movie_clean_df
```

```python
movie_sorted_ranking = movie_clean_df.sort_values('weighted_average', ascending = False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head()
```

```python
weight_average = movie_sorted_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1= sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.show()
```

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.show()
```

## Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movie_clean_df[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movie_clean_df[['normalized_weight_average','normalized_popularity']]= movie_normalized_df
```

```python
movie_clean_df.head()
```

```python
movie_clean_df['score'] = movie_clean_df['normalized_weight_average'] * 0.5 + movie_clean_df['normalized_popularity'] * 0.5
movies_scored_df = movie_clean_df.sort_values(['score'], ascending=False)
movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20)
```

```python
scored_df = movie_clean_df.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))

ax = sns.barplot(x=scored_df['score'].head(10), y=scored_df['original_title'].head(10), data=scored_df, palette='deep')

#plt.xlim(3.55, 5.25)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.show()
```
